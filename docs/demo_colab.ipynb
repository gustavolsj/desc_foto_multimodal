````python
import google.generativeai as genai
from google.colab import userdata
import PIL.Image
import json

# Configura tu API KEY de Google AI Studio
# Puedes obtener una en: https://aistudio.google.com/app/apikey
API_KEY = "TU_API_KEY_AQUI" # O usa userdata.get('GEMINI_API_KEY') si está en secretos de Colab

genai.configure(api_key=API_KEY)
model = genai.GenerativeModel('gemini-1.5-flash')
````

## 1. Definir el Prompt de Catalogación
Este es el mismo prompt utilizado en la aplicación web para asegurar consistencia en los resultados.

````python
PROMPT = """
Analiza esta fotografía histórica y devuelve un objeto JSON con la siguiente estructura:
{
  "epoca_estimada": "década o año aproximado",
  "autor_probable": "nombre o 'Anónimo'",
  "ubicacion_estimada": "ciudad, país o región",
  "coordenadas": { "lat": 0.0, "lng": 0.0 },
  "descripcion": "narrativa breve de lo que ocurre en la imagen",
  "descriptores": ["etiqueta1", "etiqueta2"],
  "confianza": 0-100,
  "justificacion": "explicación técnica de por qué se asignan estos datos basándose en elementos visuales (uniformes, edificios, tecnología, etc). Incluye enlaces de referencia si es posible",
  "descripcion_material": "análisis del soporte (ej: gelatina de plata, albúmina, daguerrotipo) según el aspecto de la imagen"
}
"""
````

## 2. Cargar y Analizar una Imagen
Puedes cargar un archivo local o usar una URL.

````python
# Para probar, descarga una imagen de muestra o sube una al panel lateral
# !wget https://example.com/foto_historica.jpg -O sample.jpg

def analizar_foto(path):
    img = PIL.Image.open(path)
    response = model.generate_content([PROMPT, img])
    
    # Limpiar la respuesta si Gemini devuelve markdown (```json ... ```)
    text = response.text.replace('```json', '').replace('```', '').strip()
    return json.loads(text)

# Ejecución (ajusta la ruta al archivo que subas)
# resultado = analizar_foto("tu_foto.jpg")
# print(json.dumps(resultado, indent=2, ensure_ascii=False))
````
